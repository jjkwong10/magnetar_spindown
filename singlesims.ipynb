```python
from magnetar import Magnetar
import matplotlib.pyplot as plt
```

## Single Magnetar Simulations

```python
Magnetari = Magnetar(M_NS=1.4,R_NS=10,gamma=1,delta=1,decay_a=0,decay_alpha=1.25,B0=10E+15,
                         v_m=26500000,rho_m=1/6.022E+23,t0=0,
                         P0=0.1,t_obvs=10E+19)
    
Magnetari.Integratedipole()
    
if Magnetari.obvs_type == '0':
    Magnetari.Integrateprop()

Magnetari.ProdNumerOutput()
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1,3)
fig.set_figheight(6)
fig.set_figwidth(18)

ax1.plot(Magnetari.t_arr, Magnetari.P_arr, ds="steps-mid",color='darkgreen', alpha=0.5,label='unbinned')
ax2.plot(Magnetari.t_arr, Magnetari.Omega_arr, ds="steps-mid", color='darkgreen',label='binned')
ax3.plot(Magnetari.P_arr, Magnetari.P_dot_arr, ds="steps-mid", color='darkgreen',label='binned')
ax1.set_xscale('log')
ax1.set_yscale('log')
ax2.set_xscale('log')
ax2.set_yscale('log')
ax3.set_xscale('log')
ax3.set_yscale('log')



ax1.set_xlabel('Frequency [Hz]')
ax1.set_ylabel('Power [arbitrary units]')

ax2.set_xscale('log')


ax2.set_xlabel('Frequency [Hz]')
ax2.set_ylabel('Power [arbitrary units]')


ax3.set_xscale('log')

ax3.set_xlabel('Frequency [Hz]')
ax3.set_ylabel('Power [arbitrary units]')
```
