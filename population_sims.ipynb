```python
from magnetar import Magnetar
import numpy as np
import pandas as pd
from tqdm import tqdm
from astropy.table import Table
from astropy.io import ascii
```

## Random Distributions

```python
Input = pd.read_csv('Input Files/magnetar_population_model_input.csv')
Input
```

```python
B0_rand = Input['B0'].to_numpy()
P0_rand = Input['P0'].to_numpy()
v_m_rand = Input['v_m'].to_numpy()

t_obvs_rand = Input['t_obvs'].to_numpy()
```

## Population Simulations

```python
population_info = []
for i in tqdm(range(len(t_obvs_rand))):
    Magnetari = Magnetar(M_NS=1.4,R_NS=10,gamma=-1,delta=0,decay_a=0,decay_alpha=1.25,B0=B0_rand[i],
                         v_m=v_m_rand[i],rho_m=1/6.022E+23,t0=0,
                         P0=P0_rand[i],t_obvs=t_obvs_rand[i])
    
    Magnetari.Integratedipole()
    
    if Magnetari.obvs_type == '0':
        Magnetari.Integrateprop()
        
    Magnetari.ProdNumerOutput()
    obvs_info = {'B0': Magnetari.B_arr[0],
                'v_m': Magnetari.v_m.value,
                'rho_m': Magnetari.rho_m.value,
                'P0': Magnetari.P_arr[0],
                'Pdot_0': Magnetari.P_dot_arr[0],
                't_obvs': (Magnetari.t_obvs*u.s).to(u.yr).value,
                'obvs_type': Magnetari.obvs_type,
                'P_obvs': Magnetari.P_arr[-1],
                'Pdot_obvs': Magnetari.P_dot_arr[-1],
                'B_obvs': Magnetari.B_arr[-1],
                'distance': Magnetari.distance.value,
                'Accretion kT_eff': Magnetari.kT_eff.value
                    }
    population_info.append(obvs_info)
    
    Status = [{'iteration': int(i),
                'B0': Magnetari.B_arr[0],
                'v_m': Magnetari.v_m.value,
                'rho_m': Magnetari.rho_m.value,
                'P0': Magnetari.P_arr[0],
                'Pdot_0': Magnetari.P_dot_arr[0],
                't_obvs': (Magnetari.t_obvs*u.s).to(u.yr).value,
                'obvs_type': Magnetari.obvs_type,
                'P_obvs': Magnetari.P_arr[-1],
                'Pdot_obvs': Magnetari.P_dot_arr[-1],
                'B_obvs': Magnetari.B_arr[-1],
                'distance': Magnetari.distance.value,
                'Accretion kT_eff': Magnetari.kT_eff.value
                    }]
    ascii.write(Table(Status), 'magnetar_population_model_status_A.csv', 
                format='csv', overwrite=True)
    
ascii.write(Table(population_info), 'Output Files/magnetar_population_model_output_A.csv', 
            format='csv', overwrite=True)
```
